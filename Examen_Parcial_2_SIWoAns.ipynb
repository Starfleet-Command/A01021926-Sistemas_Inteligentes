### <font color='blue'>P5.- ¿Cuáles son los tipos de aprendizaje máquina?. Además proporciona la diferencia entre ellas.</font>

(Valor: 3 Puntos)

---

```python
Los tipos que mas vimos en clase son el supervisado y el no supervisado, pero también existe un tercero, el de refuerzo.
El aprendizaje supervisado es aquel en el cual se conoce cada estado al que debe de llegar el modelo, y se puede retroalimentar al mismo en cada fase del entrenamiento para la correccion de errores
En el aprendizaje no supervisado, en cambio, no se conoce el estado final, y por lo tanto no se retroalimenta al modelo en su entrenamiento.
Finalmente, en el aprendizaje de refuerzo, se conoce el estado final, pero no los intermedios, y se basa en múltiples iteraciones de escenarios para llegar a la manera óptima de llegar al destino
```

(Valor: 6 Puntos)

---

```python
El gradiente descendiente es un método para llegar a un óptimo local mediante el calculo de la pendiente de una función, y el movimiento de el algoritmo en el espacio de la funcion dependiendo de la pendiente de la misma
Así, se busca tomar 'pasos' en diferentes direcciones con el objetivo de llegar al óptimo mínimo de la función, los cuales son proporcionales a el negativo de la pendiente. 
El ejemplo clásico es un hombre bajando de la montaña.En este escenario el hombre quiere llegar a la parte menos alta de la montaña lo antes posible, por lo que buscará los lugares de la montaña con
mayor inclinacion (pendiente negativo) y tomará pasos para bajar en ellos. El hombre puede terminar en un mínimo local (como un lago en la montaña) pero en muchas ocasiones llegará hasta abajo de manera rápida
```

### <font color="blue">P8.- Describe las tres características generales de los conjuntos de datos: dimensionalidad, dispersión y resolución.</font>

(Valor: 3 Puntos)

---

```python
La dimensionalidad es la cantidad de atributos o elementos que tiene cada dato que conforma al conjunto, la cual puede variar entre 1 y varios miles. Entre mayor dimensionalidad, mayor complejidad computacional, 
y mayor necesidad de obtener cuales atributos son los realmente importantes/mas significativos.

La dispersión es que tan alejados están los datos de sus medidas de tendencia central, como la media. Datos muy dispersos pueden significar poca correlación, mientras que datos muy cercanos pueden significar overfitting

La resolución, o granularidad, es el detalle de los datos: la frecuencia de su recolección, la unidad de medida, y el nivel de agregacion son todos factores que afectan la resolución de los datos. Este atributo
ayuda a determinar el 'nivel' óptimo en el cual los datos pueden ofrecer conclusiones útiles: datos de muy alta resolución se pueden usar para predecir cosas día con día o a nivel local, por ejemplo. 
```

## Predecirán el precio de las bitcoins por meses...
### Verificación de estacionariedad y descomposición STL de la serie

```python
plt.figure(figsize=[15,7])
sm.tsa.seasonal_decompose(df_month.Weighted_Price).plot()
print("Dickey–Fuller test: p=%f" % sm.tsa.stattools.adfuller(df_month.Weighted_Price)[1])
plt.show()
```

```python
df_month['Weighted_Price_box'], lmbda = stats.boxcox(df_month.Weighted_Price)
print("Dickey–Fuller test: p=%f" % sm.tsa.stattools.adfuller(df_month.Weighted_Price)[1])
```

```python
df_month['prices_box_diff'] = df_month.Weighted_Price_box - df_month.Weighted_Price_box.shift(12)
print("Dickey–Fuller test: p=%f" % sm.tsa.stattools.adfuller(df_month.prices_box_diff[12:])[1])
```

```python
# Regular differentiation
df_month['prices_box_diff2'] = df_month.prices_box_diff - df_month.prices_box_diff.shift(1)
plt.figure(figsize=(15,7))

# STL-decomposition
sm.tsa.seasonal_decompose(df_month.prices_box_diff2[13:]).plot()   
print("Dickey–Fuller test: p=%f" % sm.tsa.stattools.adfuller(df_month.prices_box_diff2[13:])[1])

plt.show()
```

```python
# Initial approximation of parameters using Autocorrelation and Partial Autocorrelation Plots
plt.figure(figsize=(15,7))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(df_month.prices_box_diff2[13:].values.squeeze(), lags=48, ax=ax)
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(df_month.prices_box_diff2[13:].values.squeeze(), lags=48, ax=ax)
plt.tight_layout()
plt.show()
```

```python
# Initial approximation of parameters
Qs = range(0, 2)
qs = range(0, 3)
Ps = range(0, 3)
ps = range(0, 3)
D=1
d=1
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)

# Model Selection
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')
for param in parameters_list:
    try:
        model=sm.tsa.statespace.SARIMAX(df_month.Weighted_Price_box, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 12)).fit(disp=-1)
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
```

```python
# Best Models
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
print(best_model.summary())
```

```python
# Inverse Box-Cox Transformation Function
def invboxcox(y,lmbda):
    if lmbda == 0:
        return(np.exp(y))
    else:
        return(np.exp(np.log(lmbda*y+1)/lmbda))
```
